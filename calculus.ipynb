```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# generating 100 linearly spaced numbers
x = np.linspace(-4,4,100)

# defining our function y = x^2
y = x**2

# plotting the function
plt.plot(x,y,'r')

# showing the plot
plt.show()
```

The following functions have important graphs to recognize:

1. $y = x$
2. $y = x^2$
3. $y = x^3$
4. $y = e^x$
5. $y = log(x)$
6. $y=1$

```python
# generating 100 linearly spaced numbers
x = np.linspace(-10,10,100)

# setting the axes at the centre
fig,axs = plt.subplots(2,3,figsize=(15,10))

# defining our functions
y = x**2
z = x**3
g = x
h = [math.exp(item) for item in x]
j = [math.log(item) for item in np.linspace(1,10,50)]
a = np.ones(100)

for ax in axs.flat:
    ax.spines['left'].set_position('center')    
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.set_ylim(-10,10)
    ax.set_xlim(-10,10)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('y=x^2')
axs[0, 1].plot(x, z, 'tab:orange')
axs[0, 1].set_title('y=x^3')
axs[1, 0].plot(x, g, 'tab:green')
axs[1, 0].set_title('y=x')
axs[1, 1].plot(x,h, 'tab:red')
axs[1, 1].set_title('y = e^x')
axs[0, 2].plot(x[50:],j, 'tab:blue')
axs[0, 2].set_title('y = log(x)')
axs[1, 2].plot(x,a, 'tab:purple')
axs[1, 2].set_title('y = 1')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.8, 
                    wspace=0.4, 
                    hspace=0.4)


# showing the plot
plt.show()
```
